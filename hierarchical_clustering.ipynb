# <b>Simple Hierarchical Clustering Example</b>

```python
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
```

```python
#create a simple dataset
x = np.array([[1, 1], [1.5, 1], [3, 3], [4, 4], [3, 3.5], [3.5, 4]])
```

```python
#visualize the dataset
plt.scatter(x[:, 0], x[:, 1], s=50);
```

```python
#calculate distances between data points with Linkage function
linkage_matrix = linkage(x, "single")
```

```python
#create dendogram
dendogram = dendrogram(linkage_matrix, truncate_mode='none')
```

# <b>Market Segmentation with Hierarchical Clustering Example</b>

```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.cluster.hierarchy as cluster_algorithm
from sklearn.cluster import AgglomerativeClustering
```

```python
#read data from csv
shopping_data = pd.read_csv('./datasets/shopping_data.csv')
data = shopping_data.iloc[:, 3:5].values
data
```

```python
plt.figure(figsize=(10, 7))
plt.title('Market Segmentation Dendogram')
dendogram = cluster_algorithm.dendrogram(cluster_algorithm.linkage(data, method="ward"))
```

Optimum number of clusters can be obtained on dendogram.
1. Determine the largest vertical distance that does not intersect any og the other clusters.
2. Draw a horizontal line at the top and at the bottom.
3. Count the number of vertical lines goint through the herizontal line. That is the optimal K value for clusters.

For this dataset, optimal K number is 5.

```python
#cluster the data with AgglomerativeClustering function
cluster = AgglomerativeClustering(n_clusters=5, affinity="euclidean", linkage="ward")
cluster.fit_predict(data)
```

```python
#visualize clustered data
plt.figure(figsize=(10, 7))
plt.scatter(data[:, 0], data[:, 1], c=cluster.labels_, cmap="rainbow")
plt.title('Market Segmentation')
plt.xlabel("Income")
plt.ylabel("Affinity");
```
