```python
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
from sklearn import datasets
import numpy as np
```

```python
#create a dataset
X,y = datasets.make_moons(n_samples=1500, noise=.05)
```

```python
#define feature variables
x1 = X[:,0]
x2 = X[:,1]
```

```python
#visualize the dataset
plt.scatter(x1,x2,s=5)
plt.show()
```

```python
#create DBSCAN model and fit
dbscan = DBSCAN(eps=0.1)
dbscan.fit(X)
y_pred = dbscan.labels_.astype(np.int64)
```

```python
#visualize dataset with DBSCAN clustering predictions
colors = np.array(['#ff0000', '#00ff00'])
plt.scatter(x1,x2,s=5,color=colors[y_pred])
plt.show()
```

```python
#create K-Means model and fit for comparison
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
y_pred = kmeans.labels_.astype(np.int64)
```

```python
#visualize dataset with K-Means clustering predictions
colors = np.array(['#ff0000', '#00ff00'])
plt.scatter(x1,x2,s=5,color=colors[y_pred])
plt.show()  
```
